## Merging Dataframes of Transcript for WER Calculation

```python
import os
import pandas as pd
```

```python
base_dir = '/Users/cogsci-lasrlab1/Desktop/MTAA_recording'
```

```python
partic_dfs = []
```

```python
for partic in os.listdir(base_dir):
    if not partic.endswith(('.py', '.DS_Store', '.csv', '.ipynb', 'transcription', 'checkpoints')):
        partic_dir = os.path.join(base_dir, partic)
        
        partic_csv = f'{partic_dir}/{partic}_sauce/{partic}_sauce.csv'

        # check if csv exists
        if os.path.exists(partic_csv):
            df_partic = pd.read_csv(partic_csv)
            partic_dfs.append(df_partic)
            print(f'Participant {partic} added to dataframe')
        else:
            print(f'CSV {partic} does not exist')

if partic_dfs: # checks to see if we've added any dfs to it
    combined_df = pd.concat(partic_dfs, ignore_index = True)
    combined_df = combined_df.drop(columns=['Unnamed: 0'])
else:
    print('No CSV files found')
```
