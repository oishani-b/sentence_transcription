### Imports for Jiwer and Pandas

```python
import jiwer
from jiwer import wer
import pandas as pd
import os
import ast
```

### Getting Dataframes for sentence info and merged participants

```python
sent_info = pd.read_csv('sentence_info.csv')
sent_info.head(8)
```

```python
partic_df = pd.read_csv('partic_transcripts.csv')
partic_df.head(8)
```

### Merging the two dataframes to start wer processing

```python
partic_df['sentence_id'] = partic_df['Sentence ID']
partic_df = partic_df[['file_name', 'participant_id', 'sentence_id', 'Iteration', 
                       'probability', 'Expected Transcription', 
                       'large', 'base', 'medium', 'small', 'tiny']]
partic_df.head()
```

```python
df_merged = partic_df.merge(sent_info, on="sentence_id", how="left")

# reordering and renaming columns
df_merged = df_merged[['file_name', 'participant_id', 'sentence_id', 'Iteration', 
                       'probability', 'reference', 'expected_transcription',
                       'acceptable_transcription', 'large', 'base', 'medium', 
                       'small', 'tiny']]
df_merged.head(8)
```

```python
# issue - acceptable_transcription is not actually list, but string
# pre-process this to fix it

# turn the strings that look like lists into actual lists
def parse_string_lists(df, col):
    df = df.copy()
    df[col] = df[col].apply(
        lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
    return df
```

```python
df_merged_final = parse_string_lists(df_merged, 'acceptable_transcription')
df_merged_final.head()
```

```python
# sanity check to make sure all the lists are lists, or NaN float values
# in the acceptable_transcription column
df_merged_final['acceptable_transcription'].apply(type).value_counts()
```

```python
# fill out acceptable transcriptions so it is the main working column
# add a list with one item for each of these, if no alternatives
def fill_acceptable_transcription(df, expected_col, acceptable_col):

    # acceptable col is NaN, so fill it out
    mask = df[acceptable_col].isna()
    df.loc[mask, acceptable_col] = df.loc[mask, expected_col].apply(lambda x: [str(x)])

    
    # acceptable col is not NaN, so add expected col value to existing list
    df.loc[~mask, acceptable_col] = df.loc[~mask].apply(lambda x: x[acceptable_col] +
                                                        [str(x[expected_col])],
                                                       axis=1)
    return df
```

```python
df_filled = fill_acceptable_transcription(df_merged_final, 
                                          'expected_transcription', 
                                          'acceptable_transcription')
df_filled.head(8)
```

### Jiwer preprocessing functions

```python
# standardize string inputs to make more accurate wer counts
def pre_jiwer_standardize(any_string):
    wer_standardize = jiwer.Compose(
    [
        jiwer.ToLowerCase(),
        # jiwer.ExpandCommonEnglishContractions(),
        # jiwer.RemoveKaldiNonWords(),
        jiwer.RemoveWhiteSpace(replace_by_space=True),
        jiwer.RemoveMultipleSpaces(),
        jiwer.Strip(),
        jiwer.RemovePunctuation()
    ]
    )
    any_string = str(any_string).replace("neighbourhood", "neighborhood")

    return wer_standardize(any_string)
```

```python
# applying transformation for preprocessing prompt/transcript columns
def transform_col_text(df, col_name, new_col_name):
    df_copy = df.copy()
    df_copy.loc[:,new_col_name] = df_copy[col_name].apply(pre_jiwer_standardize)
    return df_copy
```

### Running jiwer preprocessing for model transcriptions

```python
# test
df_large_preprocessed = transform_col_text(df_filled,
                                           'large',
                                           'large_preproc')
df_base_preprocessed = transform_col_text(df_large_preprocessed,
                                           'base',
                                           'base_preproc')
df_medium_preprocessed = transform_col_text(df_base_preprocessed,
                                           'medium',
                                           'medium_preproc')
df_small_preprocessed = transform_col_text(df_medium_preprocessed,
                                           'small',
                                           'small_preproc')
df_tiny_preprocessed = transform_col_text(df_small_preprocessed,
                                           'tiny',
                                           'tiny_preproc')

df_preproc_final = df_tiny_preprocessed
df_preproc_final.head()
```

### Modifying WER from jiwer to work with list of strings from acceptable transcription

```python
# original wer function - not used
# creating a new wer getting row-wise error rates for transformed strings in col1, col2
def wer_transformed(df, wer_col, col1, col2):
    df_copy = df.copy()
    df_copy.loc[:, wer_col] = df.apply(lambda x:
                                  wer(x[col1], x[col2]), axis=1)
    return df_copy
```

```python
# first function: get a list of wer by comparing each option in column
# acceptable transcription to the model column

def wer_list(df, acceptable_list_col, transcript_col, wer_list_col):
    df_copy = df.copy()
    df_copy.loc[:, wer_list_col] = df.apply(lambda x: 
                                       [wer(opt, x[transcript_col]) for 
                                        opt in x[acceptable_list_col]],
                                       axis = 1)
    return df_copy    
```

```python
df_large_wer_list = wer_list(df_preproc_final,
                       'acceptable_transcription', 
                       'large_preproc', 
                       'large_wer_list')
df_large_wer_list.head()
```

```python
def min_wer(df, wer_list_col, min_wer_col):
    df_copy = df.copy()
    df_copy.loc[:, min_wer_col] = df_copy[wer_list_col].apply(min)
    return df_copy
```

```python
df_min_large_wer = min_wer(df_wer_list, 'large_wer_list', 'large_wer_min')
df_min_large_wer.head()
```
