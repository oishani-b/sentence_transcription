```python
from pathlib import Path
import shutil
import re
from collections import defaultdict

# =========================
# 1) CONFIG: set MAIN path
# =========================
MAIN = Path("/path/to/MAIN")  # <-- CHANGE THIS
S1 = MAIN / "S1"
S3 = MAIN / "S3"

# If you want a dry run first (no files copied), set to True
DRY_RUN = True

# =========================
# 2) Helpers
# =========================
SAUCE_RE = re.compile(r"_sauce_(\d+_\d+)_", re.IGNORECASE)

def expected_sentence_ids():
    # 1_1 ... 100_3 (100*3 = 300)
    return {f"{i}_{j}" for i in range(1, 101) for j in (1, 2, 3)}

EXPECTED = expected_sentence_ids()

def extract_sentence_id(filename: str):
    m = SAUCE_RE.search(filename)
    return m.group(1) if m else None

def list_participant_dirs(root: Path):
    # participant folders named t1, t2, ...
    return sorted([p for p in root.iterdir() if p.is_dir() and re.fullmatch(r"t\d+", p.name)])

def build_id_to_paths(folder: Path):
    """
    Returns: dict[sentence_id] -> list[Path] for .wav files in 'folder' that match sauce_<id>
    """
    mapping = defaultdict(list)
    for wav in folder.glob("*.wav"):
        sid = extract_sentence_id(wav.name)
        if sid:
            mapping[sid].append(wav)
    return mapping

# =========================
# 3) Main copy logic
# =========================
participants_in_s1 = list_participant_dirs(S1)
if not participants_in_s1:
    raise FileNotFoundError(f"No participant folders (t1, t2, ...) found under: {S1}")

S3.mkdir(parents=True, exist_ok=True)

summary = []

for s1_part in participants_in_s1:
    participant = s1_part.name
    
    originals_part = MAIN / participant      # e.g., MAIN/t2
    if not originals_part.exists():
        print(f"[WARN] Missing originals folder for {participant}: {originals_part} (skipping)")
        continue
    
    s3_part = S3 / participant               # e.g., MAIN/S3/t2
    s3_part.mkdir(parents=True, exist_ok=True)

    # What exists in S1 participant folder?
    s1_map = build_id_to_paths(s1_part)
    present_ids = set(s1_map.keys())

    missing_ids = sorted(EXPECTED - present_ids, key=lambda x: (int(x.split("_")[0]), int(x.split("_")[1])))

    # Index originals so we can fetch missing ids quickly
    orig_map = build_id_to_paths(originals_part)

    copied = 0
    not_found = []

    for sid in missing_ids:
        candidates = orig_map.get(sid, [])
        if not candidates:
            not_found.append(sid)
            continue

        # If multiple candidates exist, we copy all of them (safe default).
        # If you only want 1, pick candidates[0].
        for src in candidates:
            dst = s3_part / src.name
            if dst.exists():
                continue  # already there
            if DRY_RUN:
                # show what would happen
                pass
            else:
                shutil.copy2(src, dst)
            copied += 1

    summary.append((participant, len(present_ids), len(missing_ids), copied, len(not_found)))

    print(
        f"{participant}: present={len(present_ids)} "
        f"missing_ids={len(missing_ids)} copied_files={copied} not_found_ids={len(not_found)}"
    )
    if not_found:
        print("  Not found in originals (first 15):", not_found[:15])

print("\nDone.")
print("DRY_RUN =", DRY_RUN)
print("If this looks right, set DRY_RUN = False and re-run.")
```
