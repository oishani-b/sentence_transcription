```python
step1_t4_path = '/Users/cogsci-lasrlab1/Desktop/MTAA_recording/Step1.Listen through these audios/t4'
step3_t4_path = '/Users/cogsci-lasrlab1/Desktop/MTAA_recording/Step3.Audios missing from Step 1/t4'

s1_t4_files = [k for k in os.listdir(step1_t4_path) if 
                k != '.DS_Store']
s1_t4_sent_ids = set()
for i in s1_t4_files:
    s1_t4_sent_ids.add(i.split('_')[6])

s1_t4_missing_ids = set()
for j in all_numbers:
    if j in all_numbers and j not in s1_t4_sent_ids:
        s1_t4_missing_ids.add(j)

#print(s1_t4_sent_ids)
#print(all_numbers)

# these are the missing ids in step 1 for participant t4
print(s1_t4_missing_ids)
```

```python
# first, we need to find the corresponding wav files from our folder
base_folder = '/Users/cogsci-lasrlab1/Desktop/MTAA_recording'

base_t4_folder = os.path.join(base_folder, 't4')

base_t4_files = [k for k in os.listdir(base_t4_folder) 
                       if len(k.split('_')) >= 7 and k.endswith('.wav')]

relevant_t4_list = [i for i in base_t4_files 
                    if i.split('_')[6] in s1_t4_missing_ids]
# print(relevant_t4_list)

# then we need to copy these over to the correct step3 folder
for relevant_file in relevant_t4_list:
    src_path = os.path.join(base_t4_folder, relevant_file)
    destin_path = os.path.join(step3_t4_path, relevant_file)

    if os.path.exists(src_path):
        with open(src_path, 'rb') as src:
            with open(destin_path, 'wb') as dst:
                dst.write(src.read())
```
